```python
from langdetect import detect
def detect_lang(s):
    i = 0
    d = {'de':0, 'en':0}
    while i <30:
        l=detect(s)
        if l=='fr' or l=='ca':
            return 'fr'
        elif l=='en':
            d['en'] += 1
        elif l=='de':
            d['de'] += 1
        i += 1
    if d['en'] == 0 and d['de'] == 0:
        return 'na'
    elif d['en'] >= d['de']:
        return 'en'
    else:
        return 'de'

# detect_lang('Grand Stylet Ergonomique Bleu Gamepad Nintendo Wii U - Speedlink Pilot Style')
df.loc[:20,'lang2'] = df.loc[:20,'designation'].apply(detect_lang)
df.head(20)
```

```python
from google.cloud import translate

def detect_language(text: str) -> translate.DetectedLanguage:
    client = translate.TranslationServiceClient()

    response = client.detect_language(content=text)

    return response.languages[0]

print(detect_language('Grand Stylet Ergonomique Bleu Gamepad Nintendo Wii U - Speedlink Pilot Style'))
```
