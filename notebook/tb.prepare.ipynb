```python
import pandas as pd
# chargement des csv de base
df = pd.read_csv('../data/X_train.csv', index_col=0)
target = pd.read_csv('../data/Y_train.csv', index_col=0)
df2 = df.join(target)

# nettoyage caractères spéciaux
import html
def convert_html_characters(s):
    if isinstance(s, str):
        return html.unescape(s)
    else:
        return s
df2['description'] = df2['description'].apply(convert_html_characters)

# suppression des balises html
from bs4 import BeautifulSoup
def extract_text_from_html(html):
    if isinstance(html, str):
        soup = BeautifulSoup(html, 'lxml')  # You can also use 'html.parser'
        return soup.get_text()
    else:
        return html
df2['description_text']=df['description'].apply(extract_text_from_html);

```

```python
# Identification des langues
# Example text
# text = "Dies ist ein Beispieltext in deutscher Sprache."
# Detect the language
# language = detect(text)
# print("Detected language:", language)

#test=df2[df2.prdtypecode == 10].copy()
#test.loc[:,'lang'] = test['designation'].apply(detect_lang)
#test.head(50)
from langdetect import detect
def detect_lang(s):
    return detect(s)
df2.loc[:,'lang'] = df2['designation'].apply(detect_lang)

```

```python
#sauvegarde result
df2.to_csv('../data/df.csv')
```
