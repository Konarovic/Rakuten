```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data/jc.df.csv')
# concaténation designation / description
df['text'] = df.apply(lambda row: row.designation + ' ' + str(row.description) if row.description != np.nan else row.designation, axis=1)
df.head()
```

```python
from langdetect import detect_langs
def detect_language_with_langdetect(line): 
    try: 
        langs = detect_langs(line) 
        return langs
        # for item in langs: 
            # The first one returned is usually the one that has the highest probability
            # return item.lang, item.prob 
    except: return "err", 0.0 

print(detect_language_with_langdetect(df.iloc[2]['text']))
```
